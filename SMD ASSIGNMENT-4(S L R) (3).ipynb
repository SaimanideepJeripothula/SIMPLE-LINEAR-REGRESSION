```python
# importinng required libraries
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
# importing dataset
data=pd.read_csv("delivery_time.csv")
```

```python
# Display the data
data
```

```python
# renaming the column names" As the varibales are nor appropriate"
dataset=data.rename({"Delivery Time":"delivery_time","Sorting Time":"sorting_time"},axis=1)
dataset
```

# EDA and DATA VISUALISATION & CORRELATION

```python
data.corr()
```

```python
# scatterplot that shows the association between 
plt.plot(dataset.delivery_time,dataset.sorting_time, "bo")
plt.xlabel("sorting_time")
plt.ylabel("delivery_time")
plt.show()
```

```python
sns.regplot(x=dataset['sorting_time'],y=dataset['delivery_time'])
```

```python
plt.figure(figsize=(5,5))
sns.distplot(dataset.delivery_time)
plt.show()
plt.figure(figsize=(5,5))
sns.distplot(dataset.sorting_time)
plt.show()
```

# Data information

```python
dataset.info()
```

# Building model

```python
import statsmodels.formula.api as smf
model=smf.ols("delivery_time~sorting_time",data=dataset).fit()
model.summary()
```

```python
#Square Root transformation on data
model1 = smf.ols('np.sqrt(delivery_time)~np.sqrt(sorting_time)',data=dataset).fit()
model1.summary()
```

```python
#Cuberoot transformation on Data
model2 = smf.ols('np.cbrt(delivery_time)~np.cbrt(sorting_time)',data=dataset).fit()
model2.summary()
```

```python
#Log transformation on Data
model3 = smf.ols('np.log(delivery_time)~np.log(sorting_time)',data=dataset).fit()
model3.summary()
```

```python
import statsmodels.api as smf
smf.qqplot(np.log(dataset['delivery_time']), line = 'r')
plt.title('No transformation')
smf.qqplot(np.sqrt(dataset['delivery_time']), line = 'r')
plt.title('Log transformation')
smf.qqplot(np.sqrt(dataset['delivery_time']), line = 'r')
plt.title('Square root transformation')
smf.qqplot(np.cbrt(dataset['delivery_time']), line = 'r')
plt.title('Cube root transformation')
plt.show()
```

```python
new_data=pd.Series([5,7,8])
newdata_pred=pd.DataFrame(new_data,columns=["sorting_time"])
```

```python
newdata_pred
```

# Making predictions using model

```python
model.predict(newdata_pred)
```

```python
#Log transformation on Data
model3 = smf.ols('np.log(Salary)~np.log(YearsExperience)',data=data1).fit()
model3.summary()
```
